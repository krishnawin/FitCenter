```python
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision
import cv2 as cv
import pandas as pd
import os
```

```python
model_path= './pose_landmarker_heavy.task'
```

```python
BaseOptions = mp.tasks.BaseOptions
PoseLandmarker = mp.tasks.vision.PoseLandmarker
PoseLandmarkerOptions = mp.tasks.vision.PoseLandmarkerOptions
VisionRunningMode = mp.tasks.vision.RunningMode
```

```python
cap = cv.VideoCapture("./archive/lateral raise/lateral raise_17.mp4")
fps = cap.get(cv.CAP_PROP_FPS)
width = int(cap.get(cv.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv.CAP_PROP_FRAME_HEIGHT))

# Initialize video writer
fourcc = cv.VideoWriter_fourcc(*"mp4v")
out = cv.VideoWriter("./", fourcc, fps, (width, height))
```

```python
options = PoseLandmarkerOptions(
    base_options=BaseOptions(model_asset_path=model_path),
    running_mode=VisionRunningMode.VIDEO)
```

```python
def analyse_video(filepath,outfile):
    with PoseLandmarker.create_from_options(options) as landmarker:
      df = pd.DataFrame()
      cap = cv.VideoCapture(filepath)
      fps = cap.get(cv.CAP_PROP_FPS)
      width = int(cap.get(cv.CAP_PROP_FRAME_WIDTH))
      height = int(cap.get(cv.CAP_PROP_FRAME_HEIGHT))
      while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        mp_image = mp.Image(image_format=mp.ImageFormat.SRGB, data=frame)
        pose_landmarker_result = landmarker.detect_for_video(mp_image, int(cap.get(cv.CAP_PROP_POS_MSEC)))
        #print(pose_landmarker_result)
        x= [landmark.x for landmark in pose_landmarker_result.pose_world_landmarks[0]]
        y= [landmark.y for landmark in pose_landmarker_result.pose_world_landmarks[0]]
        z= [landmark.z for landmark in pose_landmarker_result.pose_world_landmarks[0]]
        df = pd.concat([df,pd.concat([pd.DataFrame(x).T,pd.DataFrame(y).T,pd.DataFrame(z).T],axis=1)],ignore_index=True)
    df.to_csv(outfile)
    cap.release()
```

```python
for root, dirs, files in os.walk('../data/videos'):
    for name in files:
        filepath = root + os.sep + name
        if filepath.endswith(".mp4"):
            outpath = '../data/MediaPipe/Train/'+filepath.split('/')[-2]
            outfile = outpath+'/'+name.replace('.mp4','.csv')
            if os.path.isfile(outfile):
                continue
            if not os.path.isdir(outpath):
                os.mkdir(outpath)

            try:
                analyse_video(filepath,outfile)

            except:
                print("Error processing ", filepath)
 
```
